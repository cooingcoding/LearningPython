```python
np.random.seed(0)

df = pd.DataFrame(np.random.randn(100,3),
                 index = pd.date_range('2018-1-1', periods = 100),
                 columns = list('ABC')).cumsum()
df.head(10)
```

```python
df.plot()

plt.title('Pandas의 내장된 plot 메서드 사용 예')
plt.xlabel('날짜')
plt.ylabel('Date')
# '-' 글자가 깨짐
plt.rcParams['axes.unicode_minus'] = False # 축 음수표시 오류 해결
plt.show()
```

- method plot에 kind인수를 이용하여 다양한 시각화를 할 수 있다.
    - bar, ple, hist, kde, box, scatter 등

```python
import seaborn as sns

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

###  막대그래프(1)

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title('iris의 꽃받침의 길이를 이용한 시각화') #그래프의 제목
plt.xlabel('data')
plt.ylabel('꽃받침의 길이')
plt.show()

# rot = 0 => x축 숫자 돌리기(rotation)
```

### 막대그래프(2)

```python
iris[:5].plot.bar(rot = 0)
plt.title('Iris 붓꽃 데이터의 막대그래프')
plt.xlabel('Data')
plt.ylabel('각 feature의 값')
plt.ylim(0,7)
#plt.legend(loc = 1) ->  위치지정
plt.legend('가나다라')
plt.show()
```

####  품종별로 각 특성의 평균값

```python
df2 = iris.groupby(iris.species).mean() # 변수로 지정 = 데이터프레임이름.특성으로 묶기().평균
df2
```

```python
df2.plot.bar(rot = 0)
plt.title('품종별 feature의 평균')
plt.xlabel('품종')
plt.ylabel('품종별 평균')
```

#### 객실등급별 승개수의 비율

```python
df3 = titanic.pclass.value_counts()
df3
```

```python
df3.plot.pie(autopct = '%1.2f%%')
plt.axis('equal')
plt.show()
```
