## 바 차트(bar chart)
- bar
- barh
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh


```python
y = [2, 3, 1]
x = np.arange(len(y))

plt.title('Bar Chart')
#라벨줄거야
xlabel = ['힘', '을', '내']
plt.bar(x, y)
#xticks는 눈금이다~그 눈금에 들어가는 글자 넣을거얌
plt.xticks(x, xlabel)
#y축은 높이를 나타내주는 거라 정렬(sorted)할거야
plt.yticks(sorted(y))
plt.xlabel('x축 이름')
plt.ylabel('빈도수')
plt.show()
```

```python
#난수를 만들 때는 seed값을 줘야해
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))

#막대기를 만들려면 값이 조금 커야해 데이터도 4개
data = np.random.rand(len(people)) * 10
#error값을 만들 것이다 4개 만들어야해
error = np.random.rand(len(people))

plt.title('수평막대그래프')
plt.barh(y_pos, data, xerr=error, alpha=0.4)
#xerr=error 막대기에 작대기 꽂아 둔거
#alpha막대그래프에 투명값 막대기의 색깔을 약간 투명~하게
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

```python
# 스템플롯 : 바차트와 유사하지만 폭이 없는 그래프
# 주로 이산확륙함수나 자기상관관계를 묘사할 때 사용된다.
# http://matpotlib....롸
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('Stem Plot')
plt.stem(x, np.cos(x), '-.')
plt.show()
```

## 파이 차트(원)
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie
- 카테고리별 값의 상대적인 비교를 해야할 때 주로 사용한다
- 파이 차트를 그를 때 원의 형태를 유지할 수 있도록 plt.axis('equal')명령어를 실행해야 한다.

```python
labels = ['맥주', '소주', '양주', '막걸리']
sizes = [25, 45, 10, 20]
colors = ['yellowgreen', 'lightskyblue','gold',  'lightcoral']
#explode
explode = (0, 0.1, 0, 0)
plt.title('선생님이 마시고싶은 주류')
plt.pie(sizes, explode=explode, labels=labels,
        colors=colors, autopct='%.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

# 히스토그램
- 특정 데이터의 빈도수를 막대 모양으로 시각화한 그래프
- 도수 분포표와 히스토그램은 가장 많은 사용되는 통계 분석 도구로 데이터 특성이나 분포를 파악
- hist 명령을 사용하여 시각화하고, bins인수로 데이터를 집계할 구간 정보를 받는다.
- 가로축에는 계급이, 세루축에는 해당 도수 또는 비율을 지정한다.
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist

```python
#먼저 seed값 설정
np.random.seed(0)
x = np.random.randn(1000)
plt.title('Histogram')
arrays, bins, patches = plt.hist(x, bins=10)
plt.show()
```

# scatter plot: 산점도(산포도)
 - 2차원 데이터 즉, 두 개의 연속형(실수) 데이터 집합의 상관관계를 파악할 때 유용한 시각화 도구
 - 점 하나의 위치(x, y)의 값으로 점이 만들어 진다.
 - http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
#홍역같아서 이상하구만
```

```python
# 버블 차트(bubble chart) :  2차원이 아니라 3차원 또는 4차원의 산점도를 표현할 경우 점 하나의 크기 혹은 
# 색깔을 이용하여 다른 데이터값을 표시한다. (s- 크기, c-색깔)
N = 30
np.random.seed(0)

x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
#뒤에 뭐라고 하시는지 하하..
```

## Imshow()  : image(화상)data를 표현하는 시각화 도구
 - 이미지 데이터도 행과 열을 가진 2차원 데이터이다.
 - 

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits
#이건 이미지 데이터라 하나를 뜯어와서 볼거야
```

```python
X = digits.images[0]
X
#하나의 숫자를 만들어내는 위치
#옛날 티비는 도트로 그림을 표현
#얘는 사이즈를 지정해 놓고
# 이 위치값을 읽어오는 것
```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
#표준스칼라 메핑값이다. cmap=plt.cm.bone_r
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
#나타내는 값에 빈공축을 주어 삐죽삐죽한 것을 없애는 것
plt.xticks([])
plt.yticks([])
#눈금 표시 안되게 하기 위해false처리
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
#표준스칼라 메핑값이다. cmap=plt.cm.bone_r
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
#나타내는 값에 빈공축을 주어 삐죽삐죽한 것을 없애는 것
plt.xticks([])
plt.yticks([])
#눈금 표시 안되게 하기 위해false처리
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
methods = ['none', 'nearest', 'bilinear',
           'bicubic', 'spline16', 'spline36',
           'hanning', 'hamming', 'hermite',
           'kaiser', 'quadric', 'catrom',
           'gaussian', 'bessel', 'mitchell',
           'sinc', 'lanczos']
#그래프를 3행6열로 분할
#figsize도화지 크기
fig, axes = plt.subplots(3, 6, figsize = (12, 6),
                        subplot_kw={'xticks':[],
                                   'yticks':[]})

#왜 반복하지?
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, interpolation=interp_method, cmap=plt.cm.bone_r)
    ax.set_title(interp_method)
    
plt.show()
#보간법에 따라 숫자가 나온다
```

## Seaborn
- http://seaborn.pydata.org/installing.html
- Seaborn은 matplotlib을 기반으로 다양한 색상과 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지
- 기본적인 시각화 기능은 Matplotlib package에 의존
- 통계기능은 Statsmodels 패키지에 의존

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# 붓꽃 데이터
iris = sns.load_dataset("iris") 
# 타이타닉호 데이터
titanic = sns.load_dataset("titanic")
# 여객운송 데이터
flights = sns.load_dataset("flights")
# 팁 데이터
tips = sns.load_dataset("tips")
```

```python
iris
```

```python
#1차원 실수 플롯
# rugplot은 데이터위치를 x축 위에 작은 선분(rug)으로
##나타내어 실제 데이터의 위치를 보여준다.
#petal_width의 값을 쓸거야
x = iris.petal_width.values
sns.kdeplot(data=iris, x="petal_width")

#rug는 데이터의 위치만 나타내주는 것이다.
plt.ylim(0.0, 1.0)
sns.rugplot(x)
#plt.title("Iris 데이터 중, 꽃잎의 너비에 대한 Rug Plot ")
plt.show()
```

```python
sns.displot(x, kde=True, rug=True)
#커널 밀도를 주고 싶다.ked
#rug를 담고 싶으면 
plt.title("Iris 데이터 중, 꽃잎의 너비에 대한 Dist Plot")
plt.show()
```

```python
sns.countplot(x="class", data=titanic)
plt.title("타이타닉호의 각 클래스별, 승객 수")
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind ='kde')
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue = 'species', markers=['o', 's', 'D'])
#품종은 색상으로 넣음 hue#범주를 주고 싶을 때 
#hue : 카테고리 변수를 지정하여 카테고리에 따라 색상을 다르게 지정
#관측값마다 다르게 주고 싶다. #무슨 차이가 있는지 모르겠네...힝구
plt.show()
#이거 쓰면 좋대 예제 샘플
```

```python
titanic_size = titanic.pivot_table(
    index="class", columns="sex", aggfunc="size")
titanic_size
```

```python
# 2dimensions category type data : hearmap(in seaborn)
sns.heatmap(titanic_size,
            cmap = "crest", annot=True, fmt="d")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)

plt.show()
```

```python
#바이올렛은 카테고리도 볼 수 있다..?무슨말
#zjsjfdml 
sns.violinplot(x="day", y="total_bill", data=tips)

plt.show() #오징어 같다 오징어 순대 먹고 싶다
```

```python
sns.stripplot(x="day", y="total_bill", hue = "sex", data=tips)
#지터는 가로에 겹치는 것을 방지해준다고 하는데 썩~개운치않아
plt.legend(loc='best')
plt.show()
#hue를 주니 색이 바뀌면서 아몬드 빼빼로처럼 나옴 배고파
```

```python
sns.swarmplot(x="day", y="total_bill", hue = "sex",data=tips)
#모양이 해조류 같다 말미잘스럽네 미역국 먹고싶다.삶은 소라랑
plt.legend(loc=8)
plt.show()
#hue는 뭐다?색상이다 카테고리 변수를 지정하여 카테고리 따라 색상을 다르게 지정
```

```python
sns.boxplot(x = 'tip', y = "day", data = tips)
##여기에다가 점을 합쳐서 표현하기도해
sns.stripplot(x='tip', y='day', data=tips, jitter=True)

plt.show()
```

```python
#바이올렛은 카테고리도 볼 수 있다..?무슨말
#zjsjfdml 
sns.violinplot(x="day", y="total_bill", data=tips)
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.show() #오징어 같다 오징어 순대 먹고 싶다
```
