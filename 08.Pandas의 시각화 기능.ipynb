- Pandas의 시리즈나 데이터프레임의 plot이라는 메서드가 내장되어 있다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
#멧플롯 한글화
mpl.rc('font',family='D2coding') ## family에 원하는 코드(D2코딩)
mpl.rc('font',family='Malgun Gothic') ## (맑은 고딕)
```

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('2018-1-1', periods=100),
                   columns=list('ABC')).cumsum()
#누적합 cumsum()
df.head()
```

```python
df.plot() #인덱스가 x축, 값이 value

plt.title('Pandas의 내장 된 plot 메서드 사용 예')
plt.xlabel('날짜')
plt.ylabel('Data')
plt.show()
```

- plot 메서드에 kind인수를 통해 다양한 플롯을 시각화 할 수 있다.
- bar, pie, hist, kde, box, scatter

```python
import seaborn as sns

iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
```

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title("iris의 꽃받침의 길이 시각화")
plt.xlabel("data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5]
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 붓꽃 데이터의 막대 그래프")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 10)
plt.legend(loc=1)
plt.show()
```

```python
#품종별로 각 특성의 평균을 시각화하시오.
df2 = iris.groupby(iris.species).mean()

df2
```

```python
df2.plot.bar(rot=0)
plt.title("품종별 Feature별 평균")
plt.xlabel("species")
plt.ylabel("mean")
plt.ylim(0, 7)
plt.legend(loc=2)
plt.show()
```

```python
#위를 바꿀 때 T!
df2.T.plot.bar(rot=0)
plt.title("품종별 Feature별 평균")
plt.xlabel("species")
plt.ylabel("mean")
plt.ylim(0, 7)
plt.legend(loc=1)
plt.show()
```

```python
#객실등급별 승객수의 비유을 시각화
df3 = titanic.pclass.value_counts()
df3 #이런 결과가 나와야지, 시각화 하기가 좋다
```

```python
df3.plot.pie(autopct='%1.2f%%') #소수점 둘재자리 까지
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.show()
```

```python
iris.plot.box()
plt.show()
```
