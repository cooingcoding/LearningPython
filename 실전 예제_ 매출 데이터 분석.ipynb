```python
#가장 많이 주문한 메뉴 5개 출력
item_count = chipo['item_name'].value_counts()[:5]
for idx, (val,cnt) in enumerate(item_count.iteritems(), 1):
    print("인기메뉴", idx, ":", val, cnt)
```

```python
# 메뉴별 주문 수
item_quantity = chipo.groupby('item_name')['quantity'].sum()
item_quantity[:10]
```

```python
#아이템별 주문 개수와 총량(인덱스로 나누기)
order_count = chipo.groupby('item_name')['quantity'].count().sort_index()
order_count
```
